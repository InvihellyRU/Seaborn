Столбчатая диаграмма (barplot)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.barplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, estimator='mean', errorbar=('ci', 95), n_boot=1000, seed=None, units=None, weights=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, hue_norm=None, width=0.8, dodge='auto', gap=0, log_scale=None, native_scale=False, formatter=None, legend='auto', capsize=0, err_kws=None, ci=<deprecated>, errcolor=<deprecated>, errwidth=<deprecated>, ax=None, **kwargs)

Основные параметры функции sns.barplot():

1) data: DataFrame или массив, содержащий данные для построения графика;
2) x: название переменной или вектор данных для оси X;
3) y: название переменной или вектор данных для оси Y;
4) hue: название переменной или вектор данных, определяющий расцветку столбцов;
5) order: упорядоченный список уникальных значений для оси X (если нужно задать определённый порядок столбцов);
6) hue_order: упорядоченный список уникальных значений для переменной hue;
7) estimator: функция для вычисления значений по категориям. По умолчанию np.mean. Можно использовать такие функции, как np.sum, np.median и т.д.;
8) errorbar: функция или параметр для вычисления ошибки. Поддерживает значения sd (стандартное отклонение), ci (интервал доверия), pi (предсказательный интервал) и пользовательские функции;
9) color: цвет столбцов;
10) palette: палитра цветов для переменной hue;
11) saturation: степень насыщенности цвета столбцов. Значение по умолчанию = 0.75;
12) width: задает относительную ширину столбцов. Значение по умолчанию = 0.8;
13) dodge: логическое значение, определяющее, будут ли столбцы разных категорий hue расположены рядом (если True) или наложены друг на друга (если False).

```python
penguins = sns.load_dataset('penguins').dropna()

penguins.head()
```

```python
sns.barplot(data=penguins, x='species', y='body_mass_g', 
            hue='species',
            palette='cool',
            width=0.5,
            estimator='std',
            errorbar=None,
            saturation=0.8,
            #fill=False
            edgecolor='k')

plt.grid(alpha=0.2);
```

```python
sns.barplot(data=penguins, x='species', y='body_mass_g', 
            hue='sex',
            palette='turbo',
            width=0.4,
            errorbar=None,
            saturation=0.8,
            order=['Gentoo', 'Chinstrap', 'Adelie'],
            dodge=False,
            alpha=0.7)

plt.grid(alpha=0.2);
```

```python
sns.barplot(data=penguins, x='body_mass_g', y='species', 
            hue='sex',
            palette='viridis',
            width=0.6,
            errorbar=None,
            saturation=0.8,
            hue_order=['Female', 'Male'],
            order=['Gentoo', 'Chinstrap', 'Adelie'])

plt.grid(alpha=0.2);
```
