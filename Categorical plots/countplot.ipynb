countplot

Это тип графика, который используется для визуализации количества наблюдений в каждой категории переменной.
График выглядит как столбчатая диаграмма, где по одной оси указаны категории, а по другой - количество наблюдений в каждой категории. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.countplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, hue_norm=None, stat='count', width=0.8, dodge='auto', gap=0, log_scale=None, native_scale=False, formatter=None, legend='auto', ax=None, **kwargs)

Основные параметры функции sns.countplot():

1) data: DataFrame или массив, содержащий данные для построения графика;
2) x: название переменной или вектор данных для оси X;
3) y: название переменной или вектор данных для оси Y;
4) hue: название переменной или вектор данных, определяющий расцветку столбцов;
5) order: упорядоченный список уникальных значений для той оси, на которой отображаются категориальные данные;
6) hue_order: упорядоченный список уникальных значений для переменной hue;
7) color: цвет столбцов;
8) palette: палитра цветов для переменной hue;
9) saturation: степень насыщенности цвета столбцов. Значение по умолчанию = 0.75;
10) width: задает относительную ширину столбцов. Значение по умолчанию = 0.8;
11) dodge: логическое значение, определяющее, будут ли столбцы разных категорий hue расположены рядом (если True) или наложены друг на друга (если False);
12) stat: определяет, какую статистику нужно вычислять и отображать на диаграмме ('count', 'frequency', 'density', 'percent');
13) formatter: функция форматирования значений.

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
my_palette = ['Aquamarine', 'LightSalmon', 'MediumSlateBlue', 'BurlyWood', 'SpringGreen', 
              'HotPink', 'MediumSeaGreen']

sns.countplot(data=diamonds, x='color',
              hue='color',
              palette=my_palette,
              width=0.7,
              linewidth=1,
              edgecolor='grey',
              hatch='/');

# Основное предназначение функции sns.countplot() - это работа с категориальными данными, 
# а функции sns.histplot() - работа с числовыми значениями.
```

```python
sns.countplot(data=diamonds, x='cut',
              palette='turbo',
              hue='color');
```

```python
sns.countplot(data=diamonds, x='color',
              hue='color',
              palette='terrain',
              order=diamonds['color'].value_counts(ascending=True).index);
```

```python
plt.rcParams['hatch.linewidth'] = 7

sns.countplot(data=diamonds, x='color', 
              color='MediumSeaGreen',
              edgecolor='SeaGreen',
              hatch='x');
```

```python
plt.rcParams['hatch.linewidth'] = 8

sns.countplot(data=diamonds, x='color', 
              hue='color',
              edgecolor='grey',
              hatch='o',
              palette='terrain');
```
