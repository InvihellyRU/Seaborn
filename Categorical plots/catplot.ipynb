catplot

Функция sns.catplot() (categorical plot) служит универсальным инструментом для создания различных типов графиков, предназначенных для визуализации категориальных данных. С использованием функции sns.catplot() можно построить такие диаграммы, как: barplot, boxplot, violinplot, boxenplot, stripplot, swarmplot, pointplot и countplot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.catplot(data=None, *, x=None, y=None, hue=None, row=None, col=None, kind='strip', estimator='mean', errorbar=('ci', 95), n_boot=1000, seed=None, units=None, weights=None, order=None, hue_order=None, row_order=None, col_order=None, col_wrap=None, height=5, aspect=1, log_scale=None, native_scale=False, formatter=None, orient=None, color=None, palette=None, hue_norm=None, legend='auto', legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, ci=<deprecated>, **kwargs)

sns.catplot()

Основные параметры функции:

1) data: DataFrame, массив;
2) x, y: переменные для построения графика. X и Y задают координаты;
3) hue: позволяет добавлять цветовое различие для дополнительной переменной;
4) kind: тип графика. Может принимать значения: 'strip' (по умолчанию), 'swarm', 'box', 'violin', 'boxen', 'point', 'bar' и 'count';
5) hue_order: порядок отображения категорий по параметру hue;
6) order: порядок отображения категорий;
7) estimator: функция, используемая для агрегирования данных. По умолчанию используется 'mean' (среднее значение);
8) n_boot: количество выборок для вычисления доверительного интервала с помощью бутстрапа. По умолчанию n_boot=1000;
9) seed: задает начальное значение для генератора случайных чисел, что обеспечивает повторяемость результатов при многократном запуске кода;
10) errorbar: контролирует способ отображения ошибок на графике. По умолчанию errorbar=('ci', 95);
11) row: название столбца для разбивки графика на несколько строк;
12) col: название столбца для разбивки графика на несколько столбцов;
13) row_order: определяет порядок категорий для рядов многопанельного графика;
14) col_order: определяет порядок категорий для столбцов многопанельного графика;
15) col_wrap: определяет максимальное количество столбцов в одном ряду;
16) height: высота каждого подграфика (по умолчанию height=5);
17) aspect: соотношение ширины и высоты каждого подграфика (по умолчанию aspect=1);
18) sharex: позволяет разделить ось X между несколькими подграфиками (по умолчанию sharex=True);
19) sharey: позволяет разделить ось Y между несколькими подграфиками (по умолчанию sharey=True);
20) margin_titles: используется для управления отображением заголовков графиков в многопанельных визуализациях (по умолчанию 21) margin_titles=False).

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.catplot(data=tips, x='tip', 
            color='lightgreen')

sns.set_style('whitegrid');
```

```python
sns.catplot(data=tips, x='tip', 
            kind='box', 
            color='orange');
```

```python
sns.set_style('white')

sns.catplot(data=tips, x='day', y='tip',
            palette='viridis',
            kind='boxen',
            hue='sex')

plt.grid();
```

```python
sns.catplot(data=tips, x='day', y='tip',
            palette='turbo',
            kind='point',
            hue='sex',
            dodge=0.2,
            capsize=0.1,
            estimator=np.std);
```

```python
sns.catplot(data=tips, x='day', y='tip',
            hue='day',
            palette='OrRd',
            kind='strip',
            col='time',
            size=5,
            alpha=0.5,
            aspect=0.6);
```

```python
sns.catplot(data=tips, x='day', y='total_bill',
            palette='RdBu',
            hue='day',
            kind='box',
            col='time', 
            col_order=['Dinner', 'Lunch'], 
            row='sex', 
            row_order=['Male', 'Female'],
            aspect=0.8);
```

```python
sns.catplot(data=tips, x='day',
            color='LightSalmon',
            kind='count',
            col='size',
            col_wrap=3,
            width=0.5,
            height=3);
```

```python
sns.catplot(data=tips, x='day', y='tip',
            hue='day',
            palette='magma',
            col='time',
            row='sex',
            height=3,
            facet_kws={'despine': False});
```

```python
sns.catplot(data=tips, x='day', y='tip',
            hue='day',
            palette='Pastel1',
            col='time',
            row='sex',
            height=4,
            facet_kws={'despine': False},
            sharex=False,
            sharey=False);
```
