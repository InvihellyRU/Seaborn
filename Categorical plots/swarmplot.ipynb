swarmplot

Вид диаграммы рассеяния, который используется для визуализации распределения данных, обеспечивая, чтобы точки не накладывались друг на друга. Это достигается путем распределения точек на определенные позиции на оси, чтобы каждая точка была видна.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.swarmplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor=None, linewidth=0, hue_norm=None, log_scale=None, native_scale=False, formatter=None, legend='auto', warn_thresh=0.05, ax=None, **kwargs)

sns.swarmplot()

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.swarmplot(tips['total_bill']);
```

```python
my_palette = ['Salmon', 'grey', 'Turquoise', 'royalblue']

sns.swarmplot(data=tips, x='day', y='total_bill',
              hue='day',
              palette=my_palette,
              legend=False);
```

```python
sns.swarmplot(data=tips, x='day', y='tip',
              hue='sex',
              palette=my_palette);
```

```python
sns.swarmplot(data=tips, x='day', y='tip',
              hue='smoker',
              palette='Spectral',
              dodge=True);
```

```python
sns.boxplot(data=tips, x='day', y='total_bill',
            color='Cornsilk');

sns.swarmplot(data=tips, x='day', y='total_bill',
              hue='day',
              palette=my_palette,
              alpha=0.7,
              legend=False);
```

```python
sns.violinplot(data=tips, x='day', y='total_bill',
               color='Cornsilk',
               density_norm='count',
               inner=None);

sns.swarmplot(data=tips, x='day', y='total_bill',
              hue='day',
              palette=my_palette,
              size=4,
              legend=False);
```

```python
sns.swarmplot(data=tips, y='total_bill', 
              color='MediumTurquoise',
              linewidth=1);
```

```python
sns.swarmplot(data=tips, y='total_bill', 
              linewidth=2,
              color='white',
              edgecolor='coral');
```

```python
sns.swarmplot(data=tips, y='total_bill',
              color='HotPink',
              marker='d');
```
