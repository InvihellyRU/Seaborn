pointplot

Функция sns.pointplot() используется для построения точечных графиков, которые визуализируют категориальные данные. Этот вид графиков отображает различные статистические показатели (средняя, медиана, стандартное отклонение и пр.) и их доверительные интервалы для различных категорий.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.pointplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, estimator='mean', errorbar=('ci', 95), n_boot=1000, seed=None, units=None, weights=None, color=None, palette=None, hue_norm=None, markers=<default>, linestyles=<default>, dodge=False, log_scale=None, native_scale=False, orient=None, capsize=0, formatter=None, legend='auto', err_kws=None, ci=<deprecated>, errwidth=<deprecated>, join=<deprecated>, scale=<deprecated>, ax=None, **kwargs)#

sns.pointplot()

Основные параметры функции sns.pointplot():

1) data: DataFrame, массив;
2) x, y: переменные для построения графика. X и Y задают координаты;
3) hue: позволяет добавлять цветовое различие для дополнительной переменной;
4) hue_order: порядок отображения категорий по параметру hue;
5) dodge: логическое значение или число, определяющее сдвиг точек для разных значений категориальной переменной;
6) order: порядок отображения категорий;
7) estimator: функция, используемая для агрегирования данных. По умолчанию используется 'mean' (среднее значение);
8) n_boot: количество выборок для вычисления доверительного интервала с помощью бутстрапа. По умолчанию n_boot=1000;
9) errorbar: контролирует способ отображения ошибок на графике. По умолчанию errorbar=('ci', 95);
10) capsize: длина черточек на концах доверительных интервалов. По умолчанию capsize=0;
11) markers: стиль маркеров для точек. Можно указать символ или список символов для разных значений категориальной переменной;
12) linestyles: стиль линий, соединяющих точки. Можно указать строку или список строк для разных значений категориальной переменной.

Основные опции для параметра errorbar:

1) 'ci': Отображение доверительного интервала. Формат: ('ci', значение), где значение - размер доверительного интервала в процентах. По умолчанию errorbar=('ci', 95) - отображает 95%-й доверительный интервал.
2) 'pi': Отображение предсказательного интервала. Формат: errorbar=('pi', значение), где значение - это размер предсказательного интервала в процентах.
3) 'se': Отображение стандартной ошибки. Формат: errorbar=('se', значение), где значение - множитель для стандартной ошибки (обычно 1).
4) 'sd': Отображение стандартного отклонения. Формат: errorbar=('sd', значение), где значение - множитель для стандартного отклонения.
5) None: Отключение отображения ошибок.

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
sns.pointplot(data=diamonds, x='price', y='color',
              linestyles='none',
              hue='color',
              estimator=np.std,
              palette='seismic_r',
              capsize=0.15);
```

```python
sns.pointplot(data=diamonds, x='price', y='color',
              palette='turbo',
              linestyle='none',
              estimator='median',
              hue='cut',
              dodge=0.15,
              errorbar=None,
              markersize=5)

plt.grid();
```

```python
sns.pointplot(data=diamonds, x='color', y='price',
              palette='turbo_r',
              estimator='median',
              hue='cut',
              errorbar=None);
```
