jointplot

Функция sns.jointplot() позволяет строить графики, показывающие взаимосвязь между двумя переменными, а также их отдельные распределения.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.jointplot(data=None, *, x=None, y=None, hue=None, kind='scatter', height=6, ratio=5, space=0.2, dropna=False, xlim=None, ylim=None, color=None, palette=None, hue_order=None, hue_norm=None, marginal_ticks=False, joint_kws=None, marginal_kws=None, **kwargs)

Параметры функции:

1) data: DataFrame, содержащий данные;
2) x, y: названия столбцов или массивы данных для осей X и Y, соответственно;
3) hue: указывает на категориальную переменную, которая будет использоваться для цветового кодирования точек на графике;
4) kind: тип графика для визуализации совместного распределения. Возможные значения:
    1) 'scatter': Точечный график (по умолчанию).
    2) 'hist': Гистограмма.
    3) 'reg': График регрессии с линией регрессии и доверительными интервалами.
    4) 'resid': График остатков.
    5) 'kde': График ядерной оценки плотности.
    6) 'hex': Hexbin график;
5) height: высота графика в дюймах (по умолчанию height=6);
6) ratio: отношение высоты боковых графиков к высоте основного графика (по умолчанию ratio=5);
7) space: расстояние между боковыми графиками и основным графиком (по умолчанию space=0.2);
8) dropna: если True, пропущенные значения будут исключены из данных;
9) xlim: кортеж, определяющий пределы оси X;
10) ylim: кортеж, определяющий пределы оси Y;
11) color: цвет графиков;
12) palette: палитра цветов для различных значений переменной hue;
13) hue_order: определяет порядок категорий, отображаемых на графике;
14) hue_norm: нормализует цветовую шкалу;
15) marginal_ticks: управляет отображением меток на боковых графиках (по умолчанию marginal_ticks=False);
16) joint_kws: словарь с ключевыми словами для настройки основного графика:
    1) Параметр order в словаре joint_kws используется для настройки полиномиального порядка регрессии,
       если вы строите график регрессии с помощью kind='reg'. Этот параметр позволяет указать степень полинома;
18) marginal_kws: словарь с ключевыми словами для настройки боковых графиков;
19) cmap: меняет цветовую палитку основного графика при kind='hex';


Примечание:
График остатков residplot показывает, насколько модель регрессии ошибается в предсказаниях. На оси X располагаются входные данные (то, что мы подаём в модель), а на оси Y - остатки, то есть разница между реальными значениями и предсказанными моделью.

Когда вы используете residplot, seaborn автоматически строит простую линейную регрессионную модель для данных. По сути, он пытается найти линию, которая лучше всего подходит вашим данным, и затем показывает, насколько каждое значение отклоняется от этой линии.

```python
penguins = sns.load_dataset('penguins').dropna()

penguins.head()
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
              kind='reg',
              color='lightblue');
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
              kind='hex',
              color='#A4F6CB',
              xlim=[35, 52.5],
              ylim=[13.5, 20],
              ratio=4);
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
              kind='kde',
              hue='species',
              palette='cubehelix_r');
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
              kind='reg',
              joint_kws={'marker': 'X',
                         'color': 'grey',
                         'ci':None,
                         'order':5},
              marginal_kws={'color': 'RosyBrown',
                            'fill':False});
```
