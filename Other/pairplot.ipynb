pairplot

Функция sns.pairplot() используется для создания матрицы диаграмм рассеяния. 
Она автоматически создает сетку графиков для визуализации взаимосвязей между всеми парами переменных в наборе данных и часто используется для предварительного анализа данных.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.pairplot(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None, size=None)

sns.pairplot()

Основные параметры

1) data: DataFrame, содержащий данные для визуализации;
2) hue: позволяет применять цветовую дифференциацию для различных категорий;
3) hue_order: порядок отображения категорий для переменной hue;
4) palette: палитра цветов для категорий;
5) vars: список переменных для включения в графики. Если не указано, будут использованы все числовые переменные;
6) x_vars: список переменных для отображения на оси X;
7) y_vars: список переменных для отображения на оси Y;
8) kind: тип графика для нижней и верхней треугольной части матрицы. Может быть 'scatter', 'kde', 'hist' или 'reg'. По умолчанию 'scatter';
9) diag_kind: тип графика для диагональных элементов. Может быть auto, hist, или kde;
10) markers: маркеры для категорий;
11) height: высота каждого подграфика в дюймах (по умолчанию 2.5);
12) aspect: соотношение ширины к высоте подграфика (по умолчанию 1);
13) corner: если True, строит только нижнюю часть диагонали и графики по диагонали (по умолчанию False);
14) dropna: если True, исключает пропущенные значения перед построением графиков (по умолчанию False);
15) plot_kws: позволяет передавать дополнительные аргументы для настройки графиков, отображаемых на нижнем и верхнем треугольниках матрицы;
16) diag_kws: позволяет передавать дополнительные аргументы для настройки диагональных графиков;
17) grid_kws: позволяет передавать дополнительные аргументы для настройки всей сетки графиков.

Параметр diag_kind в функции sns.pairplot() определяет тип графика, который будет отображаться на диагональных подграфиках матрицы.

Возможные значения параметра diag_kind:

1) 'auto' (по умолчанию): Seaborn автоматически выбирает тип графика в зависимости от данных. Обычно это будет гистограмма.
2) 'hist': Построение гистограмм для каждой переменной на диагональных подграфиках. Гистограмма показывает распределение значений переменной, разбивая их на интервалы (бины) и отображая количество значений в каждом интервале.
3) 'kde': Построение графиков оценки плотности ядра для каждой переменной. Эти графики дают сглаженное представление распределения данных, показывая, где значения наиболее плотны.
4) None: Если указать None, диагональные подграфики не будут построены. Вернее, они будут построены, но на них отобразится зависимость переменной от самой себя.

Параметр diag_kws позволяет передавать дополнительные параметры для настройки диагональных графиков; diag_kws принимает словарь, в котором ключи являются именами параметров, а значения - их значениями.

Параметры, которые можно передавать через diag_kws, зависят от типа диагонального графика (diag_kind), который используется: гистограмма ('hist') или оценка плотности ядра ('kde').

Некоторые параметры для гистограммы:
1) bins: Количество бинов в гистограмме.
2) fill: Если True, то будет применена заливка столбцов цветом.
3) color: Цвет столбцов гистограммы.
4) alpha: Прозрачность столбцов гистограммы.
5) edgecolor: Цвет границ столбцов гистограммы.
6) cumulative: Если True, то гистограмма будет кумулятивной.
7) linewidth: Толщина линий границ столбцов.

Некоторые параметры для kde:
1) fill: Если True, то область под кривой будет залита цветом.
2) cut: Длина расширения плотности за пределы данных (в единицах ширины полосы).
3) color: Цвет кривой.
4) alpha: Прозрачность кривой.
5) linewidth: Толщина линии кривой.

Параметр plot_kws фунции sns.pairplot() позволяет передавать дополнительные аргументы для настройки графиков, отображаемых на нижнем и верхнем треугольниках матрицы.

Некоторые параметры для kind='scatter':

1) color: Цвет точек.
2) marker: Тип маркера.
3) alpha: Прозрачность точек.
4) s: Размер точек.
5) edgecolor: Цвет границ точек.
7) linewidth: Толщина линий границ точек.

Некоторые параметры для kind='reg':

1) ci: Доверительный интервал для линии регрессии.
2) color: Цвет точек, линии регрессии и доверительного интервала.
3) scatter_kws: Словарь параметров для настройки точек scatterplot.
4) line_kws: Словарь параметров для настройки регрессионной линии.
5) marker: Тип маркера.
6) truncate: Если True, обрезает регрессионную линию по границам данных.

Параметр grid_kws функции sns.pairplot() позволяет передавать дополнительные параметры для настройки сетки подграфиков.

Аргументы grid_kws:
1) despine: Удаление верхней и правой рамок у подграфиков.
2) layout_pad: Отступы между подграфиками.
3) diag_sharey: Если True, диагональные подграфики будут иметь общую ось Y.

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.pairplot(tips, 
             kind='hist',
             plot_kws={'color': 'pink'},
             diag_kws={'color': 'purple'});
```

```python
sns.pairplot(tips,
             hue='time',
             palette='bwr_r',
             diag_kind='kde',
             hue_order=['Dinner', 'Lunch'],
             y_vars='total_bill',
             x_vars=['tip', 'size'],
             markers='^');
```

```python
sns.pairplot(tips,
             height=2,
             diag_kws={'color': 'lightgrey',
                       'bins': 20},
             x_vars=['total_bill', 'tip'],
             y_vars=['total_bill', 'tip']);
```

```python
sns.pairplot(tips,
             height=2,
             diag_kind='kde',
             diag_kws={'fill': False,
                       'cut':0,
                       'color': 'green',
                       'linewidth': 3},
             plot_kws={'color': 'purple',
                       'alpha': 0.5,
                       's':15});
```

```python
sns.pairplot(tips,
             kind='reg',
             diag_kws={'color': 'lightgrey'},
             plot_kws={'color': 'lightblue',
                       'scatter_kws': {'color': 'red',
                                       's': 20},
                       'marker': 'x'});
```

```python
sns.pairplot(tips,
             kind='reg',
             diag_kws={'color': 'lightgrey'},
             plot_kws={'color': 'Coral',
                       'scatter_kws': {'color': 'LightSeaGreen'}},
             grid_kws={'despine': False,
                       'layout_pad': 2});
```
