```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.histplot(data=None, *, x=None, y=None, hue=None, weights=None, stat='count', bins='auto', binwidth=None, binrange=None, discrete=None, cumulative=False, common_bins=True, common_norm=True, multiple='layer', element='bars', fill=True, shrink=1, kde=False, kde_kws=None, line_kws=None, thresh=0, pthresh=None, pmax=None, cbar=False, cbar_ax=None, cbar_kws=None, palette=None, hue_order=None, hue_norm=None, color=None, log_scale=None, legend=True, ax=None, **kwargs)

sns.histplot()

1) data: данные, которые мы хотим визуализировать. Обычно это датафрейм или массив;
2) x, y: название столбца данных, который мы хотим использовать для построения гистограммы (если data - это DataFrame);
3) hue: используется для добавления цвета по категории данных;
4) multiple: определяет, как будут отображаться подгруппы, указанные с помощью параметра hue;
5) bins: количество корзин (по умолчанию seaborn сам выбирает подходящее количество);
6) binwidth: определяет ширину каждого бина (корзины) в гистограмме;
7) binrange: этот параметр позволяет задавать диапазон значений, которые будут охвачены гистограммой;
8) kde: логический параметр, который определяет, строить ли поверх гистограммы плотность распределения (KDE - Kernel Density Estimate). По умолчанию принимает значение False;
9) stat: определяет, какую статистику нужно вычислять и отображать на гистограмме;
10) cumulative: используется для создания накопительной (кумулятивной) гистограммы (по умолчанию cumulative=False);
11) element: определяет, как будет визуализироваться гистограмма. Например, 'bars' для столбиков или 'step' для ступенчатой линии;
12) color: цвет гистограммы.
13) fill: заливка гистограммы цветов (по умолчанию fill=True)
14) shrink: задает ширину бинов (по умолчанию 1), где 1 - полная ширина бина

Возможные значения параметра stat:

1) 'count': Количество наблюдений в каждом бине (используется по умолчанию). Гистограмма показывает, сколько значений попадает в каждую корзину (бин).
2) 'frequency': Частота наблюдений в каждом бине. Эта статистика делит количество наблюдений в каждом бине на общее количество наблюдений, показывая долю наблюдений.
3) 'density': Указывает, что вместо количества наблюдений (то есть частоты), по оси ординат будет отображаться плотность распределения. Эта статистика нормализует гистограмму так, чтобы суммарная площадь всех баров равнялась 1. Это полезно для сравнения распределений с разным количеством данных.
4) 'probability': Вероятность наблюдений в каждом бине. Похожа на density, но вместо плотности показывает вероятность (т.е. сумма высот всех баров равна 1).
5) 'percent': Процент наблюдений в каждом бине. Эта статистика умножает частоту на 100, чтобы показать долю в процентах.
6) 'proportion': Каждый бин гистограммы показывает долю данных, которые попадают в этот бин относительно общего числа данных. То есть, сумма всех высот бинов будет равна 1.

Возможные значения параметра element:

1) 'bars' (по умолчанию): Отображает бины в виде столбцов (баров). Это традиционный вид гистограммы, где высота каждого столбца соответствует количеству наблюдений или другой выбранной метрике.
2) 'step': Отображает гистограмму в виде ступенчатого графика. В этом режиме высоты бинов соединяются прямыми линиями, создавая вид ступенек.
3) 'poly': Отображает гистограмму в виде линейного графика. В этом режиме вершины каждого бина соединяются прямыми линиями, создавая замкнутую линию, напоминающую линейный график.

Возможные значение параметра multiple:

1) 'layer' - подгруппы накладываются друг на друга полупрозрачными слоями (по умолчанию);
2) 'dodge' - размещает подгруппы бок о бок;
3) 'stack' - подгруппы складываются друг на друга, создавая суммарную высоту;
4) 'fill' - нормализует высоты так, чтобы они суммировались до 1.

```python
penguins = sns.load_dataset('penguins')
penguins.replace([np.inf, -np.inf], np.nan, inplace=True)
penguins.dropna(inplace=True)
```

```python
sns.set_style('whitegrid')
sns.histplot(x='bill_length_mm', data=penguins, 
             kde=True, 
             color='orange', 
             alpha=0.7, 
             #bins=20,
             binwidth = 1,
             binrange=(32, 55),
             lw=3,
             element='step',
             fill=False);
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='percent',
             fill=False,
             element='step',
             linewidth=4,
             cumulative=True,
             binwidth = 2);

plt.axhline(80, color='red');
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
             cbar=True,
             color='purple'
             );
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
             palette=['#DC143C', '#FFFACD', '#2E8B57'],
             #palette=['red', 'b', 'g'],
             hue='species',
             alpha=0.7);
```
