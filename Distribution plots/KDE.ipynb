Ядерная оценка плотности (Диаграмма KDE)
Диаграмма KDE (Kernel Density Estimate) строит сглаженную оценку плотности вероятности для непрерывных данных. Это позволяет визуально оценить, как значения распределены по всему диапазону данных.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.kdeplot(data=None, *, x=None, y=None, hue=None, weights=None, palette=None, hue_order=None, hue_norm=None, color=None, fill=None, multiple='layer', common_norm=True, common_grid=False, cumulative=False, bw_method='scott', bw_adjust=1, warn_singular=True, log_scale=None, levels=10, thresh=0.05, gridsize=200, cut=3, clip=None, legend=True, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)

sns.kdeplot()

1) data: DataFrame, массив с данными;
2) x: имя переменной, которая будет отображена на оси X;
3) y: имя переменной, которая будет отображена на оси Y. Если задан, строится 2D KDE;
4) hue: имя переменной для группировки данных по категориям, каждая из которых будет представлена отдельной линией KDE;
5) weights: массив значений для взвешивания точек данных. Должен быть такой же длины, как и данные;
6) palette: палитра цветов для различения категорий, заданных параметром hue;
7) hue_order: список для определения порядка категорий, отображаемых при использовании hue;
8) hue_norm: два значения, определяющие нормализацию данных по оси hue;
9) color: цвет линий и заливки, если hue не используется;
10) fill: логическое значение, определяющее, следует ли заполнять область под KDE;
11) multiple: определяет как отображать несколько KDE: 'layer', 'stack' или 'fill' (по умолчанию 'layer');
12) common_norm: логическое значение, определяющее, нормализовать ли плотности по всем категориям hue вместе (True) или отдельно (False). По умолчанию common_norm=True;
13) common_grid: логическое значение, определяющее, использовать ли общую сетку для всех категорий hue. По умолчанию common_grid=False;
14) cumulative: логическое значение, определяющее, строить ли накопительную функцию распределения (по умолчанию cumulative=False);
15) bw_method: метод для вычисления ширины окна (bandwidth): 'scott' (по умолчанию), 'silverman' или числовое значение;
16) bw_adjust: множитель для ширины окна. Умножает значение, вычисленное bw_method, на этот коэффициент (по умолчанию bw_adjust=1);
17) log_scale: Логическое значение или кортеж из двух логических значений для использования логарифмической шкалы по осям X и/или Y;
18) levels: количество уровней контуров для 2D KDE (по умолчанию levels=10);
19) thresh: минимальное значение плотности для отображения контуров в 2D KDE (по умолчанию thresh=0.05);
20) gridsize: количество точек в сетке для оценки плотности (по умолчанию gridsize=200);
21) cut: количество стандартных отклонений для увеличения границ осей при построении KDE (по умолчанию cut=3). Указывает, насколько далеко за пределы данных следует отображать кривую KDE;
22) clip: пределы значений для обрезки KDE;
23) legend: логическое значение, определяющее, отображать ли легенду (по умолчанию legend=True);
24) cbar: логическое значение, определяющее, отображать ли цветовую шкалу (по умолчанию cbar=False).
25) linewidth (lw): задает толщину линии у графика;

```python
penguins = sns.load_dataset('penguins').dropna()

penguins.head()
```

```python
sns.kdeplot(penguins['body_mass_g'],
           fill = True,
           lw = 3,
           color = '#A4F6CB',
           clip = (3000, 6000)
           )
plt.show()
```

```python
sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
            fill=True,
            cbar=True,
            levels=15)
plt.show()
```

```python
sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
            hue='species',
            fill=True,
            palette='cubehelix',
            thresh=0.2)
plt.show() 
```
