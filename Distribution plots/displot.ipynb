displot

Функция sns.displot() (distribution plot) позволяет создавать различные виды графиков распределения, такие как гистограммы, KDE-плотности и ECDF, а также поддерживает возможность разделения на подграфики для отображения данных по подгруппам.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.displot(data=None, *, x=None, y=None, hue=None, row=None, col=None, weights=None, kind='hist', rug=False, rug_kws=None, log_scale=None, legend=True, palette=None, hue_order=None, hue_norm=None, color=None, col_wrap=None, row_order=None, col_order=None, height=5, aspect=1, facet_kws=None, **kwargs)

Основные параметры функции sns.displot():

1) data: DataFrame или массив, содержащий данные для визуализации;
2) x, y: имена столбцов для построения графиков по осям X и Y;
3) hue: столбец для группировки данных по цвету;
4) col, row: столбцы для создания подграфиков;
5) kind: тип графика ('hist' для гистограммы, 'kde' для KDE-плотности, 'ecdf' для ECDF);
6) rug: добавляет штрих-диаграмму (по умолчанию - False);
7) rug_kws: позволяет увеличить высоту штрих-диаграммы;
8) bins: количество или границы бинов для гистограммы;
9) binwidth: ширина каждого бина;
10) palette: палитра цветов для группировки по hue;
11) kde: логическое значение для добавления KDE-плотности к гистограмме;
12) stat: статистика для оси Y ('count', 'density', 'probability', 'frequency', 'percent').
13) multiple: определяет, как будут отображаться подгруппы, указанные с помощью параметра hue;
14) facet_kws: отвечает за отображение верхней и правой границы подграфиков. Если вам нужно, чтобы не удалялись верхняя и правая границы подграфиков, то используйте параметр facet_kws, которому нужно передать словарь с ключом {'despine': False}.
15) height: позволяет определить высоту каждого подграфика в дюймах (по умолчанию height=5).
16) aspect: позволяет определить соотношение сторон подграфиков, т.е. отношение ширины к высоте (по умолчанию aspect=1).

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.displot(x='total_bill', data=tips, 
            binwidth=1,
            color='orange',
            kde=True,
            stat='probability')

plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
sns.displot(x='total_bill', data=tips, 
            kind='kde',
            fill=True,
            linewidth=4,
            color='green',
            rug=True,
            rug_kws={'height': 0.04})
plt.show()
```

```python
sns.displot(x='total_bill', data=tips,
            hue='time',
            multiple='stack')

plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
sns.displot(x='total_bill', data=tips,
            hue='time',
            kind='kde',
            cut=0)

plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
sns.displot(x='total_bill', data=tips,
            col='sex',
            row='time',
            hue='sex',
            palette=['#781c6d', '#A4F6CB'],
            height=2.5,
            aspect=1.5);
```

```python
sns.displot(x='total_bill', y='size', data=tips, 
            cbar=True)

plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
sns.displot(x='total_bill', y='size', data=tips, 
            kind='kde',
            rug=True)

plt.grid(linestyle='--', alpha=0.5)
plt.show()
```
