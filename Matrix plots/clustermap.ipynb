clustermap

Функция sns.clustermap() используется для создания тепловых карт с иерархической кластеризацией по строкам и/или столбцам. Эта функция используется для визуализации многомерных данных и поиска в них скрытых структур, таких как группы или кластеры.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.clustermap(data, *, pivot_kws=None, method='average', metric='euclidean', z_score=None, standard_scale=None, figsize=(10, 10), cbar_kws=None, row_cluster=True, col_cluster=True, row_linkage=None, col_linkage=None, row_colors=None, col_colors=None, mask=None, dendrogram_ratio=0.2, colors_ratio=0.03, cbar_pos=(0.02, 0.8, 0.05, 0.18), tree_kws=None, **kwargs)

sns.clustermap()

Основные параметры функции:

1) data: двумерный массив или DataFrame, который будет визуализироваться;
2) pivot_kws: позволяет передавать дополнительные аргументы в метод pivot() pandas перед построением тепловой карты;
3) method: метод кластеризации, такой как 'single', 'complete', 'average', 'weighted', 'centroid', 'median', или 'ward' (по умолчанию method='average');
4) metric: метрика расстояния, используемая для вычисления расстояний между строками и/или столбцами. По умолчанию используется евклидово расстояние metric='euclidean';
5) z_score: используется для стандартизации данных (по умолчанию z_score=None - стандартизация не производится);
6) standard_scale: используется для масштабирования данных перед их кластеризацией и визуализацией (по умолчанию standard_scale=None - масштабирование не производится);
7) figsize: размер фигуры в дюймах (по умолчанию figsize=(10, 10));
8) row_cluster: Логическое значение, определяющее, будут ли строки кластеризованы (по умолчанию row_cluster=True);
9) col_cluster: Логическое значение, определяющее, будут ли столбцы кластеризованы (по умолчанию col_cluster=True);
10) row_linkage: массив, определяющий иерархическую кластеризацию для строк (по умолчанию row_linkage=None);
11) col_linkage: массив, определяющий иерархическую кластеризацию для столбцов (по умолчанию col_linkage=None);
12) row_colors и col_colors: позволяют добавлять цветные аннотации к строкам и столбцам тепловой карты;
13) colors_ratio: используется для управления соотношением между размерами аннотаций цветов (row_colors и col_colors) и основной частью тепловой карты (по умолчанию colors_ratio=0.03);
14) mask: используется для скрытия определенных ячеек тепловой карты;
15) dendrogram_ratio: соотношение размеров дендрограммы к тепловой карте. Может быть кортежем для указания отдельных значений для строк и столбцов (по умолчанию dendrogram_ratio=0.2);
16) cbar_pos: позиция цветовой шкалы. По умолчанию cbar_pos=(0.02, 0.8, 0.05, 0.18);
17) cbar_kws: используется для настройки внешнего вида и поведения цветовой шкалы (color bar) на тепловой карте;
18) tree_kws: позволяет настраивать внешний вид и поведение дендрограмм.

Параметр standard_scale может принимать следующие значения:

1) standard_scale=0: Масштабирование выполняется по строкам. Каждая строка будет иметь значения в диапазоне от 0 до 1. Это может быть полезным, когда строки данных имеют разный диапазон значений, и вы хотите сравнивать строки независимо от их исходного диапазона.
2) standard_scale=1: Масштабирование выполняется по столбцам. Для каждого столбца значения приводятся к диапазону от 0 до 1. Это может быть полезным, когда столбцы данных имеют разный диапазон значений, и вы хотите сравнивать столбцы независимо от их исходного диапазона.
3) standard_scale=None: Масштабирование не выполняется (по умолчанию).

Параметр z_score может принимать следующие значения:

1) z_score=0: Стандартизация выполняется по строкам. Для каждой строки из значений вычитается среднее значение этой строки, а затем результат делится на стандартное отклонение этой строки.
2) z_score=1: Стандартизация выполняется по столбцам. Для каждого столбца из значений вычитается среднее значение этого столбца, а затем результат делится на стандартное отклонение этого столбца.
3) z_score=None: Стандартизация не выполняется (по умолчанию).

Oсновные методы кластеризации, которые можно использовать с параметром method:

1) single: Одинарное связывание. Минимальное расстояние между элементами кластеров.
2) complete: Полное связывание. Максимальное расстояние между элементами кластеров.
3) average: Среднее связывание. Среднее расстояние между элементами кластеров (применяется по умолчанию).
4) weighted: Взвешенное среднее связывание. Взвешенное среднее расстояние между элементами кластеров.
5) centroid: Центроидное связывание. Расстояние между центроидами кластеров.
6) median: Медианное связывание. Медианное расстояние между элементами кластеров.
7) ward: Связывание по методу Уорда (Ward's linkage). Минимизирует сумму квадратов внутри кластера.

Mетрики расстояния, которые можно использовать с параметром metric:

1) euclidean: Евклидово расстояние.
2) minkowski: Расстояние Минковского (обобщение евклидова и манхэттенского расстояний).
3) cityblock: Манхэттенское расстояние.
4) chebyshev: Расстояние Чебышева (максимальное расстояние по одной из координат).
5) seuclidean: Стандартизированное евклидово расстояние.
6) sqeuclidean: Квадрат евклидова расстояния.
7) cosine: Косинусное расстояние (или косинусное сходство).
8) correlation: Корреляционное расстояние.

```python
df = sns.load_dataset('mpg').dropna()

df.head()
```

```python
pivot = df.pivot_table(values='horsepower', index='cylinders', 
                     columns='origin', aggfunc='count', fill_value=0)
display(pivot)
```

```python
sns.clustermap(pivot,
               cmap='Reds',
               figsize=(6, 6),
               annot=True,
               fmt='d')

plt.show()
```

```python
origin_colors = df['origin'].map({'europe': 'SlateBlue', 
                                  'japan': 'yellow',
                                  'usa': 'SpringGreen'})
```

```python
sns.clustermap(df.select_dtypes(include=['number']),
               cmap='RdYlBu',
               figsize=(7, 7),
               row_colors=origin_colors)
plt.show()
```

```python
sns.clustermap(df.select_dtypes(include=['number']),
               cmap='RdYlBu',
               figsize=(7, 7),               
               row_colors=origin_colors,
               standard_scale=1)
plt.show()
```

```python
sns.clustermap(df.select_dtypes(include=['number']),
               cmap='summer',
               figsize=(7, 7),              
               row_colors=origin_colors,
               z_score=1)
plt.show()
```

```python
sns.clustermap(df.select_dtypes(include=['number']),
               cmap='PuRd_r',
               figsize=(7, 7),              
               row_colors=origin_colors,
               metric='cityblock')
plt.show()
```
