Диаграмма рассеяния (scatter plot)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn.scatterplot(data=None, *, x=None, y=None, hue=None, size=None, style=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, legend='auto', ax=None, **kwargs)

sns.scatterplot()

Основные параметры:
1) data: DataFrame или массив с данными;
2) x: название столбца или вектор данных для оси X;
3) y: название столбца или вектор данных для оси Y;
   
1) сolor: задает цвет точек
2) alpha: задает прозрачность точек (float: 0 < alpha < 1)
3) style: задает дополнительное измерение с помощью разных форм точек (маркеров) (название столбца или вектор данных для изменения стиля точек (например, разные формы));
-- С помощью параметра markers можно задать, какие формы будут использоваться для различных категорий данных. 
4) hue: задает дополнительное измерение c помощью разных цветов (название столбца или вектор данных для группировки точек по цвету);
5) hue_order: задаёт порядок отображения категорий для переменной hue;
6) hue_norm: позволяет нормализовать данные переменной hue;
7) palette: цветовая палитра для переменной hue;
8) s: задает размер точек
9) size: название столбца или вектор данных для изменения размера точек;
10) sizes: диапазон размеров для переменной size;
11) size_order: задаёт порядок отображения категорий для переменной size;
12) size_norm: позволяет нормализовать данные переменной size, задавая диапазон значений;
13) marker: задает форму точек (маркеров)
14) markers: список стилей маркеров для переменной style;
15) style_order: задаёт порядок отображения категорий для переменной style.

Аргументы для параметра marker

'o': круг;
's': квадрат;
'^': треугольник вверх;
'v': треугольник вниз;
'>': треугольник вправо;
'<': треугольник влево;
'x': крестик;
'*': звёздочка;
'D': ромб;
'H': шестиугольник.

sns.set_style() -- стиль отображения графика // Для добавление сетки написать sns.set_style('*color*grid')

```python
penguins = sns.load_dataset('penguins').dropna()

#display(penguins.head())
#display(penguins.info())
#display(penguins.shape)
```

```python
sns.scatterplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', 
                s = 30, 
                #color = 'r', 
                alpha = 0.9,
                marker = 'X',
                hue = 'island',
                hue_order = ['Dream', 'Biscoe', 'Torgersen'],
                palette = 'deep6')
plt.show()
```
